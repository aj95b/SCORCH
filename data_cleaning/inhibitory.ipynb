```python
sc.pp.filter_cells(inh, min_genes = 200)
sc.pp.filter_genes(inh, min_cells = 10)
inh
```

```python
sc.pp.highly_variable_genes(
    inh,
    n_top_genes=10000,
    subset=True,
    layer="logcounts",
    flavor="seurat_v3",
    batch_key="study",
)
```

```python
scvi.model.SCVI.setup_anndata(inh)
vae = scvi.model.SCVI(inh)
vae.train()
solo = scvi.external.SOLO.from_scvi_model(vae)
solo.train()
df = solo.predict()
df['prediction'] = solo.predict(soft = False)
df.index = df.index.map(lambda x: x[:-2])
df.index=inh.obs.index
inh.obs['doublet'] = df['prediction']
inh = inh[inh.obs.doublet == 'singlet']
inh
```

```python
inh = inh.copy()
scvi.model.SCVI.setup_anndata(inh, layer='logcounts',
                              categorical_covariate_keys=['study'],
                              #continuous_covariate_keys=["Age at Death"]
                             )
model = scvi.model.SCVI(inh)
model.train()
SCVI_LATENT_KEY = "X_scVI"

latent = model.get_latent_representation()
inh.obsm[SCVI_LATENT_KEY] = latent
sc.pp.neighbors(inh, use_rep=SCVI_LATENT_KEY)
sc.tl.umap(inh, min_dist=0.3)
sc.set_figure_params(figsize=(10, 10))
sc.pl.umap(
    inh,
    color=["subclass"],
    frameon=False,
    size=5
)
```
